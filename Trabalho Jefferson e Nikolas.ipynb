###### Entre os anos de 2013,2014 e 2015, será analisado a taxa de óbito infantil com mães de 10 a 14 anos. 

Utilizando o site <http://tabnet.datasus.gov.br/cgi/deftohtm.exe?sim/cnv/inf10br.def> Acesso em 26/11/2018. Foi possível ter referencia e dados para que o trabalho pudesse ser criado. Sendo atualizado até os anos de 2014, 2015 e 2016, nós iremos observar a taxa de óbito infantil com mulher da faixa etária de 10 a 14 anos, nas capitais do país.

Com esse trabalho, esperamos obter resultados decrescentes, ou então, acharmos uma solução para que essa situação diminuia e a população se prontifique do quadro apresentado.

Usamos para a extração dos dados o excel baixado no mesmo site, com a extensão CSV.

```python
#código para importar as bibliotecas para a montagem das tabelas, etc.
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#esse código fez com que conseguissemos criar a tabela logo em baixo.
plt.style.use('ggplot')
```

```python
#aqui nós chamamos a nossa lista do excel em CSV com os dados dos óbitos.
obi = pd.read_csv('obitos.csv')
```

```python
#montagaem da tabela utilizando as 26 capitais.
obi.head(27)
```

```python
#assim, reogarnizamos a nossa coluna para melhor entendimento dentro do próprio pandas.
obi.columns = ['Capital', '2014', '2015', '2016']
```

```python
obi.head(27)
```

### Observação
Até aqui, podemos observar que há uma grande variação. 
Há capitais em que a taxa aumentou e outras em que diminuiu. Para melhor observação, então, montaremos um gráfico de barras, assim ficará mais fácil distinguir os dados uns dos outros.

```python
#Para melhor observação e análise dos dados, iremos criar a partir dos dados então coletados um gráfico. Resultando assim em uma
#melhor visualização do nosso problema.
```

```python
a = obi[['Capital', '2014', '2015','2016']].head(27).set_index('Capital').sort_values('2014', ascending=True)
a.plot(kind='bar', figsize=(16,5), grid=False, color=['green', 'blue', 'red'], legend=True)
plt.title('Óbitos infantis nos anos de 2014, 2015 e 2016')
plt.xlabel('Capitais')
plt.ylabel('Número de óbitos')
plt.show()
```

### Resultado:
Obtemos então o gráfico com os dados das capitais do país. Entretanto é possível notar que continuamos com muita variação, portanto iremos criar um gráfico utilizando apenas os dados das capitais onde mais teve número de óbito e mais variou.

```python
#aqui nós chamamos nosso arquivo "dados", onde colocamos as 6 capitais onde o número de óbitos era mais alto e mais variou.
b= pd.read_csv('dados.csv')
a = b[['capital', '2014', '2015', '2016']].head(6).set_index('capital').sort_values('2014', ascending=True)
a.plot(kind='barh', figsize=(13,7), grid=False, color=['blue', 'green','red'], legend=True)

#Nomeamos as linhas e as colunas para maior compreensão dos dados.
plt.title('6 Capitais com maior número de óbitos')
plt.xlabel('Óbitos')
plt.ylabel('Capitais')
plt.show()
```

### Resultados:
Notamos que em 4 das 6 capitais o número de óbitos até 2016 diminuiram drásticamente, o que é uma ótima visão. Analisamos também em 2 das 6 capitais o processo foi inverso, ou seja, os números de 2016 aumentaram. Em São Paulo, como exemplo, o número aumentou radicalmente.
